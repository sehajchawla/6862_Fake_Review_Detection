# Amazon Data Exploration

This notebook is concerned with exploring some of the yelp data. We do this by creating summary statistics and visualizations to understand the spread better, and get familiar with the dataset.

```python
# importing the relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

** Amazon Pantry Exploration **

```python
# Loading up the data
df_pantry = pd.read_csv("../../data/processed/amazon_prime_pantry.csv")
df_pantry = df_pantry.drop(['Unnamed: 0'], axis=1)  # Deleting an irrelevant column
df_pantry.head(5)  # Displaying first 5 rows of the dataframe
```

```python
# writing a function to count words in the review text
def countWords(row):
    try:
        return len(list(row['reviewText'].split(" ")))
    except:
        return None

df_pantry['wordCount'] = df_pantry.apply(countWords, axis=1)  # applying the function to each row
df_pantry.head(5)  # Displaying first 5 rows of the dataframe
```

```python
# Removing nones before plotting
wordcountlist = [] 
for val in list(df_pantry['wordCount']): 
    if val == val : 
        wordcountlist.append(val) 

plt.hist(wordcountlist, bins=50)  # creating a histogram with 50 bins to visualise the distributon
plt.title("Distribution of review word count")  # adding a title
plt.xlabel("Number of words")  # adding a x label
plt.ylabel("Frequency")  # adding a y label
plt.show()
```
